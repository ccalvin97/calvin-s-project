```python
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from gensim import corpora, models
import gensim
import csv
```

```python
########################### Stage 1 ###########################
# user's input in the description box
description = input()
```

```python
########################### Stage 2 ###########################
cleaned = []
# clean and tokenise the description
raw = description.lower()
tokeniser = RegexpTokenizer(r'\w+')
tokens = tokeniser.tokenize(raw)

# create a list containing stopwords in English
stopwords = set(stopwords.words('english'))
# remove stopwords from tokens
removedS_tokens = [description for description in tokens if not description in stopwords]

# Create pstemmer of class PorterStemmer
pstemmer = PorterStemmer()
# Reduce tokens to stemmed forms
stemmed_tokens = [pstemmer.stem(description) for description in removedS_tokens]

# final cleaned document
cleaned.append(stemmed_tokens)
```

```python
########################### Stage 3 ###########################
# Bag of words: turn the cleaned document into a dictionary with id and term
dictionary = corpora.Dictionary(cleaned)

# Convert the dictionary into a matrix of words
corpus = [dictionary.doc2bow(text) for text in cleaned]
```

```python
########################### Stage 4 ###########################
# Run LDA model
lda = gensim.models.ldamodel.LdaModel(corpus, num_topics=1, id2word=dictionary, passes=20)
```

```python
########################### Stage 5 ###########################
num_topics = 1
topic_words = []
# Store the topics and keywords generated by LDA model into a list
for i in range(num_topics):
    tt = lda.get_topic_terms(i, 10)
    topic_words.append([dictionary[pair[0]] for pair in tt])
flattened = [val for sublist in topic_words for val in sublist]
print (flattened) # checking keywords list

# Return certain information if any word in the keyword list matches with the words in car dictionary
with open("car_dictionary.csv") as f_obj:
    reader = csv.reader(f_obj, delimiter=',')
    if any(item in row for row in reader for item in flattened):
        print("Congratulations! Your post will be passed on to the next stage.")
    else:
        print("Sorry. The description submitted does not seem to be relevant. Please resubmit.")
```
